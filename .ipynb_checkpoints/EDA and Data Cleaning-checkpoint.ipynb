```python
def check_discrepancies(param, reference_param, df):
    temp_df = df[df[param].isnull()]
    temp_bool_list = (temp_df[reference_param]==0 & temp_df[param].isnull())
    
    temp_df_index_list = [i for i, val in enumerate(temp_bool_list) if val==False]
    
    print(str(len(temp_df_index_list)) + ' discrepancies found')
    display(temp_df.iloc[temp_df_index_list])
    
    return 
    
```

def one_hot_encoding(df, col_name):
    col = col_name.replace(" ","")
    df = pd.get_dummies(columns=[col_name], data=df, prefix=col, drop_first=True)
    return df
